```python
import numpy as np
import matplotlib.pyplot as plt
import ot.datasets

def generate_initial_distribution(n):
    """生成初始的混合高斯分布"""
    return (0.6 * ot.datasets.get_1D_gauss(n, 20, 5) + 
            0.4 * ot.datasets.get_1D_gauss(n, 50, 5)).astype(np.float128)

def brownian_motion(samples, steps, delta):
    """模拟布朗运动"""
    for _ in range(steps):
        samples += np.random.normal(0, delta, samples.shape)
    return samples

n = 1000  # 样本数
initial_samples = generate_initial_distribution(n)

# 模拟布朗运动
steps = 50  # 布朗运动的步数
delta = 0.1  # 每一步的方差
final_samples = brownian_motion(initial_samples, steps, delta)

# 可视化初始和最终分布
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(initial_samples, bins=50, density=True)
# plt.title("初始分布")
plt.subplot(1, 2, 2)
plt.hist(final_samples, bins=50, density=True)
# plt.title("布朗运动后的分布")
plt.show()
```
